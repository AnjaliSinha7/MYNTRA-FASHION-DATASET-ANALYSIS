# **MYNTRA FASHION DATASET ANALYSIS**

IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("/content/Fashion Dataset.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

CHANGE DESCRIPTION COLUMN

```python
df["description"] = df["description"].str.replace("Orange solid Kurta with Palazzos with dupatta","5-10")
df.head()
```

BRAND DISTRIBUTION

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df,x="brand") # store the axes object into ax
ax.bar_label(ax.containers[0]) # use ax to call bar_label
plt.show()
```

RATING ANALYSIS

```python
print("\nRating Statistics:\n", df['avg_rating'].describe())

plt.figure(figsize=(8, 6))
sns.histplot(df['avg_rating'], kde=True)
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.show()
```

BRAND ANALYSIS

```python
brand_counts = df['brand'].value_counts()
print("\nBrand Counts:\n", brand_counts)
```

Top N Brands

```python
top_brands = brand_counts.head(10) # Get top 10 brands
plt.figure(figsize=(10, 6))
top_brands.plot(kind='bar')
plt.title('Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

RELATIONSHIP BETWEEN PRICE AND RATING

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='price', y='avg_rating', data=df)
plt.title('Price vs. Average Rating')
plt.xlabel('Price')
plt.ylabel('Average Rating')
plt.show()
```

DISTRIBUTION BY PRODUCT CATEGORY AND COLOR

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='products', hue='colour', data=df, order=df['products'].value_counts().index[:10]) # Show top 10 products
plt.title('Product Distribution by Color')
plt.xlabel('Product')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

CORRELATION BETWEEN NUMERIC FEATURES

```python
correlation_matrix = df[['price', 'ratingCount', 'avg_rating']].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

GROUPING AND AGGREGATION

```python
product_price_stats = df.groupby('products')['price'].agg(['mean', 'median', 'max', 'min'])
print("\nProduct Price Statistics:\n", product_price_stats)
```

SEGEMENTATION

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='price', y='avg_rating', data=df, hue='products')  # Color points by product
plt.title('Customer Segmentation (Price vs. Rating)')
plt.xlabel('Price')
plt.ylabel('Average Rating')
plt.show()
```

COLOR ANALYSIS

```python
from wordcloud import WordCloud  # InstallING wordcloud: pip install wordcloud

text = " ".join(df['colour'].dropna().astype(str))  # Combine colors into a single string
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Color Word Cloud')
plt.show()
```
